# Categorical Plots

Seaborn also helps us in doing the analysis on categorical Data points. In this secion we will discuss about

- boxplot
- violinplot
- countplot
- barplot


```python
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('F:/Python/Tips/tips.csv')
```

```python
## Count plot -Show the counts of observations in each categorical bin using bars.hg
```

```python
sns.countplot('sex',data=df)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.countplot('sex',data=df)
```

```python
ax = sns.countplot('sex',data=df,order=df['sex'].value_counts(ascending=False).index) # making it as an descening order
abs_value = df['sex'].value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0],labels=abs_value)# Providing label
```

```python
ax = sns.countplot('sex',data=df,order=df['sex'].value_counts(ascending=False).index) # making it as an descening order
abs_value = df['sex'].value_counts(ascending=False).values #Numerical value
rel_value = df['sex'].value_counts(ascending=False, normalize=True).values * 100  # Percentage value
lbls= [f'{p[0]} ({p[1]:.1f}%)' for p in zip(abs_value,rel_value)]
ax.bar_label(container=ax.containers[0],labels=lbls)# Providing label
```

# BOX plot

A box and whishker plot (some times called a boxplot) is a graph that presents information from a five-number summary

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df, palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# Categorize my data based on some other categories

sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# Violin Plot

Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df, palette='rainbow')
```

# Pickling

All pandas objects are equipped with to_pickle methods which use Python's Pickle module to save data structures to disk using the pickle format.

Lets say we have trained the data model which took 2 hours to run

each and every time when executing it need not to run 2hours , we can pickle the model 
and use it to predict different excel as a source to get the prediction

```python
df_excel.to_pickle('df_excel')
```

```python
df=pd.read_pickle('df_excel')
```
